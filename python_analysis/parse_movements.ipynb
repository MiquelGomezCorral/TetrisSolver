```python
import re

def parse_movement_to_list(text: str) -> str:
    movements = []
    genotype = None
    line_passed = False
    for line in text.split("\n"):
        if 'Genotype' in line:
            genotype = re.search(r'\[([\w]+)\]', line).group(1)


        if line_passed and '|' in line:
            numbers = line.split('|')
            movements.append([int(num) for num in numbers[:-1]])

        if '--' in line:
            line_passed = True

    return genotype, movements

def parse_movement_to_cs(genotype: str, movements: list[list[int]]):
    int_matrix = 'new int[][]'
    int_vec = 'new int[]'
    b1, b2 = "{", "}"
    ident = "        "

    res = f"{ident}{int_matrix} {b1}\n"
    for piece in movements:
        piece = [str(move) for move in piece][1:]
        res += f"{ident}    {int_vec} {b1} {', '.join(piece)} {b2},\n"

    res += f"{ident}{b2}"

    return res

def parse_bags_to_cs(bags: list[str]) -> str:
    queue = 'new Queue<TetriminoEnum>(new [] {'
    tetrimino = 'TetriminoEnum.'
    b1, b2 = "{", "}"
    ident = "        "

    res = ""
    for bag in bags:
        res += f"{ident}{queue}\n"
        res += ",\n".join([f"{ident}    {tetrimino}{piece}" for piece in bag])

        res += f"\n{ident}{b2},\n"
    

    return res
```

# All experimentss

```python
bags = []
genotypes = []
pieces = []

for exp in GA_exp_list:
    movement_text, bag = exp.load_last_movements()
    bags.append(bag)
    genotype, movements = parse_movement_to_list(movement_text)
    genotypes.append(genotype)

    parsed = parse_movement_to_cs(genotype, movements)
    pieces.append(parsed)
```

```python
print(parse_bags_to_cs(bags))
```

```python
print(',\n'.join([str(move) for move in pieces]))
```
